1. Function

```python
# define merge_sort function
def merge_sort(list):
    if len(list) <=1:
        return list
    else:
        # divide the list
        mid=len(list)//2
        left=list[:mid]
        right=list[mid:]
        
        # merge sort both lists
        left=merge_sort(left)
        right=merge_sort(right)
        sorted_list=merge(left,right)
        return(sorted_list)
        
# define merge function
def merge(left, right):
    merged=[]
    left_index, right_index=0,0
    while left_index<len(left) and right_index<len(right):
        
        if left[left_index]<=right[right_index]:
            merged.append(left[left_index])
            left_index+=1
        else:
            merged.append(right[right_index])
            right_index+=1
    merged.extend(left[left_index:])
    merged.extend(right[right_index:])
    return merged      
    
```

2. Examination

```python
import random

random_list = [random.randint(10, 100) for _ in range(20)]
sorted_list = merge_sort(random_list)
print(f"example list:{random_list}")
print(f"sorted list:{sorted_list}")
```

3. Time measure

```python
import time
import matplotlib.pyplot as plt

# time measure
input_sizes = [10000, 20000, 40000, 80000, 100000]

sorting_times = []

for size in input_sizes:
    
    random_list = [random.randint(1, 1000000) for _ in range(size)]
    start_time = time.time()
    sorted_list = merge_sort(random_list)
    end_time = time.time()
    duration = end_time - start_time
    sorting_times.append(duration)
    print(f"Time for sorting {size} numbers is {duration} seconds")
    
# plot
plt.plot(input_sizes, sorting_times, marker='o')
plt.xlabel('Input Length')
plt.ylabel('Sorting Time (seconds)')
plt.title('Merge Sort Performance')
plt.grid(True)
plt.show()



    
    
    
    
```
